```python
df = pd.read_pickle('/content/drive/MyDrive/merged_training.pkl')
df.head()
```

```python
df.shape

df.rename(columns={'text':'message','emotions':'sentiment'}, inplace='True')
```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import (sent_tokenize, word_tokenize, TreebankWordTokenizer, wordpunct_tokenize)
```

```python
def remove_special(text):
    x=''
    
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x = x + ' '
    
    return x

df['message']=df['message'].apply(remove_special)

def low(text):
  new=text.lower()
  return new
df['message']=df['message'].apply(low)
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.remove('not')
stop_words.remove('no')

stop_words.remove('nor')
df['message'] = df['message'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df.head()
```

```python

nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
def lemm(text):
  y=[]
  text=text.split()
  for i in text:
    y.append(lemmatizer.lemmatize(i))
  return y
```

```python
df['message']=df['message'].apply(lemm)
df.head()
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
df['message']=df['message'].apply(str)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split 
from sklearn.svm import LinearSVC 
from sklearn.metrics import classification_report, confusion_matrix 

from sklearn.linear_model import LogisticRegression

from keras.utils import to_categorical
from keras import models
from keras import layers
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['message'], 
                                                    df['sentiment'], 
                                                    random_state=0)
```

```python
df['sentiment'].unique()
```

```python
tfidf=TfidfVectorizer(max_df=0.8,min_df=10)
# res=df['review'].apply(tfidf.fit_transform())
train_tfidf=tfidf.fit_transform(X_train)
test_tfidf=tfidf.transform(X_test)
print(tfidf.vocabulary_)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
clf = MultinomialNB(alpha=1)
clf.fit(train_tfidf, y_train)

y_pred = clf.predict(test_tfidf)
print('Test Accuracy: ', accuracy_score(y_test, y_pred))
```

```python
def classify_sentiment(text):
    tokens = word_tokenize(text.lower())
    words = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha() and token not in stop_words]
    text_vectorized = tfidf.transform([' '.join(words)])
    prediction = clf.predict(text_vectorized)
    return prediction
```

```python
text1 ="i am feeling unwell since past 5 days"
print(classify_sentiment(text1))
```
